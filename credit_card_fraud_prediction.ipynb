```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
```

```python
df= pd.read_csv("creditcard.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
# checking missing values
df.isnull().sum()
```

```python
# checking mathematical values
df["Class"].value_counts()
```

### Data is Highly Undistributed

```python
# dividing data
legal= df[df.Class==0]
fraud= df[df.Class==1]
```

```python
legal.shape
```

```python
fraud.shape
```

```python
# checking mathematical values for legal cases
legal["Amount"].describe()
```

```python
# checking mathematical values for fraud cases
fraud["Amount"].describe()
```

```python
# checking number of fraud cases
fraud.shape[0]
```

### Taking 492 legal cases randomly to match fraud cases and making dataframe an equally distributed dataframe

```python
legal_cases= legal.sample(492)
```

```python
legal_cases["Amount"].value_counts()
```

```python
# merging into new dataframe
new_df= pd.concat([legal_cases, fraud], axis=0)
```

```python
new_df.sample(5)
```

```python
new_df.shape
```

```python
# labeling new dataframe as input and output data
X= new_df.drop(["Class"], axis=1)
Y=new_df["Class"]
```

```python
# splitting data into test and train data
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, stratify=Y, test_size=0.2, random_state=2)
```

```python
lr=LogisticRegression()
lr.fit(X_train, Y_train)
```

```python
# checking accuracy for training data
y_pred_train=lr.predict(X_train)
```

```python
accuracy_score(Y_train, y_pred_train)
```

```python
np.mean(cross_val_score(lr, X, Y, cv= 20))
```

```python
# accuracy for test data
y_pred_test= lr.predict(X_test)
```

```python
accuracy_score(Y_test, y_pred_test)
```
